```python
class User:
    all_accounts = {}

    def __init__(self, name, email, password):
        self.__name = name
        self.__email = email
        # self.__user_id = user_id
        self.__password = User.encrypt_password(password)
        User.all_accounts[email] = self

    def get_password(self):
        return self.__password

    @staticmethod
    def encrypt_password(password):
        en_pass = ""
        for i in password:
            en_pass += chr(ord(i) - 1)
        return en_pass

    @classmethod
    def match_id_and_password(cls, email, password):
        account = cls.all_accounts.get(email)
        if account is None:
            return 0
        else:
            if account.__password == User.encrypt_password(password):
                return account
            else:
                return 1

    def logout(self):
        print(f"{self.__name} has been logged out.")


class JobSeeker(User):
    total_applicants = {}
    curriculum_v={}
    def __init__(self, name, user_id, email, password):
        super().__init__(name, user_id, email, password)
        JobSeeker.total_applicants[email] = self
    # def cv(self,academic_s,expertise):
    #   l1=[]
    #   self.academic_s=academic_s
    #   self.expertise=expertise
    #   l1.extend(self.academic_s,self.expertise)
    #   curriculum_v[name]=l1
# Dictionary to store user information
users = {}

# Function to handle user login
def login():
    email = input("Enter email: ")
    password = input("Enter password: ")
    if email in users and users[email]['password'] == password:
        print("Login successful!")
        print("Welcome, {}!".format(users[email]['name']))
    else:
        print("Invalid email or password. Please try again.")
        return 0

class signup(User):
  def __init__(self):
    print("==== Signup ====")
    email = input("Enter email: ")
    if email in User.all_accounts:
        print("Email already exists. Please try again.")
        return
    name = input("Enter name: ")
    password = input("Enter password: ")
    dob = input("Enter date of birth (dd/mm/yyyy): ")
    gender = input("Enter gender: ")
    # Store user information in dictionary
    super().__init__(name,email,password)
    users[email] = {
         'name': name,
         'password': password,
         'dob': dob,
         'gender': gender
     }
    print("Signup successful!")
class Job:
    def __init__(self, title, location, salary):
        self.title = title
        self.location = location
        self.salary = salary
    def __str__(self):
        return f"Title: {self.title}\nLocation: {self.location}\nSalary: ${self.salary}\n"

class JobBoard:
    def __init__(self):
        self.jobs = []
        for i in self.jobs:
          print(i.title)
    def post_job(self, job):
        self.jobs.append(job)

    def search_jobs(self, title=None, location=None, salary=None):
        found_jobs = []
        for job in self.jobs:
            if  title.lower() != job.title.lower():
                continue
            if location.lower() != job.location.lower():
                continue
            if salary  >= job.salary:
                continue
            found_jobs.append(job)
        print(self.jobs)
        print(found_jobs)
        return found_jobs
```
